```python
ipl_dataset['new_batter']=ipl_dataset['new_batter'].fillna("no new batter")
ipl_dataset.head()
```

```python
ipl_dataset=ipl_dataset.drop(['umpire'],axis=1,errors='ignore')
ipl_dataset.columns
```

```python
ipl_dataset['next_batter']=ipl_dataset['next_batter'].fillna("No next batter")
ipl_dataset.head()
```

```python
ipl_dataset.drop_duplicates(inplace=True)
ipl_dataset.head()
```

```python
ipl_dataset= ipl_dataset.rename(columns={'Unnamed: 0': 'ID'})
ipl_dataset.head()
```

```python
#which team has won the most matches?
ipl_dataset['match_won_by'].value_counts().head(5)
```

```python
#TOP 10 players with 'player of match'
ipl_dataset['player_of_match'].value_counts().head(10)
```

```python
#Most successfull venue
ipl_dataset['venue'].value_counts()
```

```python
#Team won after toss win
ipl_dataset[ipl_dataset['toss_winner']==ipl_dataset['match_won_by']]['match_won_by'].value_counts()
```

```python
top_teams= ipl_dataset[ipl_dataset['toss_winner']==ipl_dataset['match_won_by']]
table= top_teams['match_won_by'].value_counts().reset_index()
table.columns=['Team','toss_winner_and_match_won_by']
table
```

```python
top_teams= ipl_dataset['match_won_by'].value_counts()
top_teams.head()
```

```python
#plotting top teams 
plt.figure(figsize=(18,12))
sns.barplot(x=top_teams.index, y=top_teams.values)
plt.title("TOP  TEAMS WITH MATCH WINS")
plt.xlabel("teams")
plt.ylabel("number of wins")
plt.xticks(rotation=75)
plt.show()
```

```python
# winning margin distribution
plt.figure(figsize=(18,10))
sns.histplot(ipl_dataset['win_outcome'], bins=10)
plt.title("Distribution of Win Margins")
plt.xlabel("Run Margin")
plt.xticks(ticks=[20,40,80,120,160])
plt.show()
```

```python
#Time series Analysis
matches_per_month= ipl_dataset['month'].value_counts().sort_index()
plt.figure(figsize=[12,5])
sns.lineplot(x=matches_per_month.index, y=matches_per_month.values)
plt.xlabel("month")
plt.ylabel("Number of matches")
plt.title("matches played each month")
#plt.grid(True)
plt.show()
```
